```python
# 課題：Matplotlibで、可視化した複数のグラフを1つの描画キャンパス上で同時に表示しよう

import numpy as np
import matplotlib.pyplot as plt
# プロット
fig, axes = plt.subplots(nrows=2, ncols=3)


# 折れ線グラフ
y0 = [1, 2, -5, 2]
axes[0,0].plot(y0)

# sin波グラフ
x1 = np.linspace(0, 10, 30)
y1 = 2 + 2 * np.sin(2 * x1)

# 空のグラフが表示されてしまうため削除（https://terakoya.sejuku.net/question/detail/33796）
#fig, ax = plt.subplots()

axes[0,1].plot(x1, y1, linewidth=2.0)

axes[0,1].set(xlim=(0, 10), xticks=np.arange(0, 10),
       ylim=(-1, 5), yticks=np.arange(-1, 6))

# ヒストグラム
x2 = np.random.normal(15, 5, 2000)

axes[0,2].hist(x2)

# 散布図
np.random.seed(3)
x3 = 4 + np.random.normal(0, 2, 60)
y3 = 4 + np.random.normal(0, 2, len(x3))

axes[1,0].scatter(x3, y3)
plt.show()
```
